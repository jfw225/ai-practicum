```python
def sample_mean_and_variance(tensors):
    total_sum, total_sum_of_squares = 0.0, 0.0
    N = 0.0
    
    for tensor in tensors:
        N += tensor.numel()
        total_sum += tensor.sum()
        total_sum_of_squares += (tensor ** 2).sum()


    mean = total_sum / N

    v1 = (total_sum_of_squares - 2*mean*total_sum + N*(mean**2))/ (N-1)
    v2 = (total_sum_of_squares - total_sum**2/N )/ (N-1)
    
    print(v1.item())
    print(v2.item())
    assert v1 - v2 < 1e-4
    variance = v2

    return variance, mean

a = torch.rand(100,100,100) * -1
b = torch.rand(100,100,100) * -1
c = torch.rand(100,100,100) * -1
cat = torch.cat((a, b, c)) 
print(torch.var_mean(cat)[0].item(), torch.var_mean(cat)[1].item())
sample_mean_and_variance([a,b,c])
```

```python
def get_FWHM_gaussian_kernel(fwhm, t):
    sigma = fwhm / np.sqrt(8 * np.log(2))
    ts = torch.arange(3.31*-3,3.31*4,3.31)
    gauss = torch.exp((-(ts / sigma)**2 / 2))
    k = gauss / gauss.sum()

    kernel_3d = torch.einsum('i,j,k->ijk', k, k, k)
    kernel_3d = kernel_3d / kernel_3d.sum()
#     vol_3d = F.conv3d(t.reshape(1, 1, *t.shape).float(), kernel_3d.reshape(1, 1, *kernel_3d.shape), stride=1, padding=len(k) // 2)
    vol_3d = F.conv3d(t[None,None,:,:,:].float(), kernel_3d[None,None,:,:,:], stride=1, padding=len(k) // 2)

    return torch.squeeze(vol_3d)
```
